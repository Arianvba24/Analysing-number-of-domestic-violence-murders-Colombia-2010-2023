```python
import pandas as pd
import numpy as np
import datetime
from datetime import datetime

# CSV file
# df = pd.read_csv(r"C:\Users\Cash\Downloads\archive\raw_data.csv",index_col = "Indice",low_memory = False,parse_dates = ["fecha_hecho"])

# Parquet file
df = pd.read_parquet(r"raw_data.parquet")
```

```python
df["fecha_hecho"] = pd.to_datetime(df["fecha_hecho"], format = "%d/%m/%Y")
df.info()
```

```python
#Creating index
df["Index"] = list(range(1,df.shape[0]+1))
df.set_index("Index", inplace = True)
df
```

```python
#Number of incidents by year ordered by largest to smallest

dfx = df.copy()
dfx["fecha_hecho"] = dfx["fecha_hecho"].dt.year
x = dfx.groupby("fecha_hecho")[["cantidad"]].sum().reset_index().sort_values(by = "cantidad", ascending = False)
print(f'The total of deaths is: {x["cantidad"].sum()}')
values = {"fecha_hecho": "Years of incidents", "cantidad": "Total deaths"}
dfx.groupby("fecha_hecho")[["cantidad"]].sum().reset_index().sort_values(by = "cantidad", ascending = False).rename(columns = values)
```

```python
#Number of deaths caused by sharp elements("ARMA BLANCA / CORTOPUNZANTE")
print(f'The number of deaths caused by sharp elements is: {df.loc[df["armas_medios"]== "ARMA BLANCA / CORTOPUNZANTE", "cantidad"].sum()}')
df.loc[df["armas_medios"]== "ARMA BLANCA / CORTOPUNZANTE"]
```

```python
#Number of deaths caused by shooting in states("ARMA DE FUEGO")
values = {"departamento":"State", "cantidad": "Total deaths"}
print(f'The number of deaths caused by shooting is: {df.loc[df["armas_medios"]== "ARMA DE FUEGO", "cantidad"].sum()}')
df.loc[df["armas_medios"]== "ARMA DE FUEGO"].groupby("departamento")[["cantidad"]].sum().reset_index().sort_values(by = "cantidad", ascending = False).rename(columns = values)
```

```python
#Number of deaths by state
values = {"departamento":"State", "cantidad": "Total deaths"}
df.groupby("departamento")[["cantidad"]].sum().reset_index().sort_values(by = "cantidad",ascending = False).rename(columns = values)
```

```python
#Number of deaths by shooting in CUNDINAMARCA
print(f'The number of deaths by shooting in CUNDINAMARCA is: {df.loc[(df["armas_medios"] == "ARMA DE FUEGO") & (df["departamento"] == "CUNDINAMARCA"), "cantidad"].sum()}')
df.loc[
    (df["armas_medios"] == "ARMA DE FUEGO") & 
    (df["departamento"] == "CUNDINAMARCA")
]

# df.loc[(df["armas_medios"] == "ARMA DE FUEGO") & (df["departamento"] == "CUNDINAMARCA"), "cantidad"].sum()
```

```python
#Number of deaths by age range
values = {"grupo_etario": "Age range", "cantidad":"Total deaths"}
df.groupby("grupo_etario")[["cantidad"]].sum().reset_index().sort_values(by = "cantidad", ascending = False).rename(columns = values)
```

```python
#Max amount of deaths in a day by state in 2016
values = {"departamento":"State", "cantidad": "Total deaths"}
df.groupby("departamento")[["cantidad"]].max().reset_index().sort_values(by = "cantidad", ascending = False).rename(columns = values)
```

```python
#Comparison of female deaths in 2010 and 2020
values = {
    "cantidad_x":"Total deaths 2010",
    "cantidad_y":"Total deaths 2020",
    "genero":"gender"
    }
deaths_2010= df.loc[(df["genero"] == "FEMENINO") & (df["fecha_hecho"].dt.year == 2010)].groupby("genero")[["cantidad"]].sum().reset_index()
deaths_2020= df.loc[(df["genero"] == "FEMENINO") & (df["fecha_hecho"].dt.year == 2020)].groupby("genero")[["cantidad"]].sum().reset_index()

deaths_comp= deaths_2010.merge(deaths_2020, how = "left", left_on = "genero", right_on = "genero")
deaths_comp.rename(columns = values, inplace = True)
def calculo(deaths_comp):
    value = (deaths_comp["Total deaths 2020"] / deaths_comp["Total deaths 2010"])*100
    return round(value,2)
deaths_comp["% Diference"] = deaths_comp.apply(calculo, axis = 1)
deaths_comp
```

```python
#Remove duplicates

df = df.loc[df.duplicated(subset = ["cantidad"],keep = False) & df.duplicated(keep = False)].drop_duplicates(subset = ["departamento"], keep = "last")
df
```
